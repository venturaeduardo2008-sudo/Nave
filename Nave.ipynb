# 7) Análise de consumo energético
foi solicitado a IA um dataset para a análise, com base em um prompt

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Define the number of rows
num_rows = 500

# Define the start timestamp
start_time = datetime(2023, 1, 1, 0, 0, 0)

# Generate sequential timestamps (e.g., every minute)
timestamps = [start_time + timedelta(minutes=i) for i in range(num_rows)]

# Generate data for each column based on specified ranges
internal_temp_c = np.random.uniform(18, 35, num_rows)
external_temp_c = np.random.uniform(-5, 30, num_rows)
battery_voltage_v = np.random.uniform(46, 52, num_rows)
battery_current_a = np.random.uniform(20, 120, num_rows)
battery_soc_percent = np.random.uniform(60, 100, num_rows)
battery_capacity_ah = np.random.uniform(80, 120, num_rows)
power_load_kw = np.random.uniform(5, 25, num_rows)
energy_loss_percent = np.random.uniform(2, 8, num_rows)
tank_pressure_bar = np.random.uniform(95, 145, num_rows)
estimated_autonomy_min = np.random.uniform(45, 180, num_rows)

# States
structural_integrity = np.ones(num_rows, dtype=int) # Always 1
critical_modules_status = np.ones(num_rows, dtype=int) # Always 1
telemetry_link_status = np.ones(num_rows, dtype=int) # Always 1

# Calculate energy_available_kwh
energy_available_kwh = (battery_voltage_v * battery_capacity_ah * battery_soc_percent / 100) / 1000

# Create a dictionary for the DataFrame
data = {
    'timestamp': timestamps,
    'internal_temp_c': internal_temp_c,
    'external_temp_c': external_temp_c,
    'battery_voltage_v': battery_voltage_v,
    'battery_current_a': battery_current_a,
    'battery_soc_percent': battery_soc_percent,
    'battery_capacity_ah': battery_capacity_ah,
    'energy_available_kwh': energy_available_kwh,
    'power_load_kw': power_load_kw,
    'energy_loss_percent': energy_loss_percent,
    'tank_pressure_bar': tank_pressure_bar,
    'structural_integrity': structural_integrity,
    'critical_modules_status': critical_modules_status,
    'telemetry_link_status': telemetry_link_status,
    'estimated_autonomy_min': estimated_autonomy_min
}

df_telemetry = pd.DataFrame(data)

# Determine launch_decision
def get_launch_decision(row):
    conditions = [
        18 <= row['internal_temp_c'] <= 35,
        -5 <= row['external_temp_c'] <= 30,
        46 <= row['battery_voltage_v'] <= 52,
        20 <= row['battery_current_a'] <= 120,
        60 <= row['battery_soc_percent'] <= 100,
        80 <= row['battery_capacity_ah'] <= 120,
        5 <= row['power_load_kw'] <= 25,
        2 <= row['energy_loss_percent'] <= 8,
        95 <= row['tank_pressure_bar'] <= 145,
        45 <= row['estimated_autonomy_min'] <= 180,
        row['structural_integrity'] == 1,
        row['critical_modules_status'] == 1,
        row['telemetry_link_status'] == 1
    ]
    if all(conditions):
        return 'READY'
    else:
        return 'ABORT'

df_telemetry['launch_decision'] = df_telemetry.apply(get_launch_decision, axis=1)

# Save to CSV
csv_filename = 'telemetry_data.csv'
df_telemetry.to_csv(csv_filename, index=False)

print(f"Synthetic telemetry data saved to {csv_filename}")
```

# 8) Análise com possíveis anomalias
foi solocitado a IA que adicione anomalias no dataset anterior para a análise



```python
import numpy as np

# Create a copy of the original DataFrame to introduce anomalies
df_telemetry_anomalies = df_telemetry.copy()

# Add 'anomalia_inserida' column, initialized to 'nao'
df_telemetry_anomalies['anomalia_inserida'] = 'nao'

# Determine the number of rows to insert anomalies (3%)
num_anomalies = int(len(df_telemetry_anomalies) * 0.03)

# Get random indices for anomaly insertion
anomaly_indices = np.random.choice(df_telemetry_anomalies.index, num_anomalies, replace=False)

# Define anomaly ranges
anomaly_rules = {
    'internal_temp_c': (45, 70),
    'battery_soc_percent': (5, 25),
    'tank_pressure_bar_low': (40, 80),
    'tank_pressure_bar_high': (160, 220),
    'structural_integrity': 0,
    'critical_modules_status': 0,
    'telemetry_link_status': 0
}

# Insert anomalies
for idx in anomaly_indices:
    df_telemetry_anomalies.loc[idx, 'anomalia_inserida'] = 'sim'

    # Randomly select a type of anomaly to apply to this row
    anomaly_type_keys = list(anomaly_rules.keys())
    # Ensure at least one anomaly is applied
    num_anomalies_to_apply = np.random.randint(1, len(anomaly_type_keys) + 1)
    selected_anomaly_types = np.random.choice(anomaly_type_keys, num_anomalies_to_apply, replace=False)

    for anomaly_type in selected_anomaly_types:
        if anomaly_type == 'internal_temp_c':
            low, high = anomaly_rules[anomaly_type]
            df_telemetry_anomalies.loc[idx, 'internal_temp_c'] = np.random.uniform(low, high)
        elif anomaly_type == 'battery_soc_percent':
            low, high = anomaly_rules[anomaly_type]
            df_telemetry_anomalies.loc[idx, 'battery_soc_percent'] = np.random.uniform(low, high)
        elif anomaly_type == 'tank_pressure_bar_low':
            low, high = anomaly_rules[anomaly_type]
            df_telemetry_anomalies.loc[idx, 'tank_pressure_bar'] = np.random.uniform(low, high)
        elif anomaly_type == 'tank_pressure_bar_high':
            low, high = anomaly_rules[anomaly_type]
            df_telemetry_anomalies.loc[idx, 'tank_pressure_bar'] = np.random.uniform(low, high)
        elif anomaly_type in ['structural_integrity', 'critical_modules_status', 'telemetry_link_status']:
            df_telemetry_anomalies.loc[idx, anomaly_type] = anomaly_rules[anomaly_type]

# Recalculate energy_available_kwh for the anomalous dataset
df_telemetry_anomalies['energy_available_kwh'] = \
    (df_telemetry_anomalies['battery_voltage_v'] *
     df_telemetry_anomalies['battery_capacity_ah'] *
     df_telemetry_anomalies['battery_soc_percent'] / 100) / 1000

# Recalculate launch_decision for the anomalous dataset
def get_launch_decision(row):
    conditions = [
        18 <= row['internal_temp_c'] <= 35,
        -5 <= row['external_temp_c'] <= 30,
        46 <= row['battery_voltage_v'] <= 52,
        20 <= row['battery_current_a'] <= 120,
        60 <= row['battery_soc_percent'] <= 100,
        80 <= row['battery_capacity_ah'] <= 120,
        5 <= row['power_load_kw'] <= 25,
        2 <= row['energy_loss_percent'] <= 8,
        95 <= row['tank_pressure_bar'] <= 145,
        45 <= row['estimated_autonomy_min'] <= 180,
        row['structural_integrity'] == 1,
        row['critical_modules_status'] == 1,
        row['telemetry_link_status'] == 1
    ]
    if all(conditions):
        return 'READY'
    else:
        return 'ABORT'

df_telemetry_anomalies['launch_decision'] = df_telemetry_anomalies.apply(get_launch_decision, axis=1)

# Save the anomalous DataFrame to CSV
anomaly_csv_filename = 'telemetry_data_anomalies.csv'
df_telemetry_anomalies.to_csv(anomaly_csv_filename, index=False)

print(f"Synthetic telemetry data with anomalies saved to {anomaly_csv_filename}")
```
